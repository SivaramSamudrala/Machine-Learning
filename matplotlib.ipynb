```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x=np.linspace(0.,10.,100)
y=x*x
plt.figure(figsize=(5,5))
plt.plot(x,y,label='$x^2$')
plt.title('square function')
plt.legend(loc='best')
plt.xlabel('x')
plt.ylabel('$x^2$')
```

```python
x=np.linspace(0.,10.,100)
y=np.sin(x)
plt.figure(figsize=(9,6))
plt.plot(x,y,'r--',label='sin(x)')
plt.title('sinusoidal function')
plt.legend(loc='best')
plt.xlabel('phase Angle(rad)')
plt.ylabel('sin(x)')
plt.savefig('sin_function_png',dpi=300)
```

```python
items=np.array(['coke','pepsi','fanta','maaza','mirinda'])
qty=np.array([100,85,20,30,45])
plt.bar(items,qty)
plt.title=("sales")
plt.xlabel=('Beverages')
plt.ylabel=('qty sold')
plt.show()
```

```python
plt.pie(qty,labels=items,autopct='%0if')
plt.title=('% of sales')
plt.show()
```

```python
plt.figure()
plt.subplot(2,1,1)
plt.plot(x,np.sin(x))
plt.title=('sin(x)')
plt.ylabel=('sin(x)')
plt.subplot(2,1,2)
plt.plot(x,np.cos(x))
plt.title=('cos(x)')
plt.ylabel=('cos(x)')
plt.xlabel=('x')
```

```python
plt.figure()
fig,ax=plt.subplots(2,2)
plt.tight_layout()
ax[0,0].plot(x,np.sin(x))
ax[0,0].set_title('sin(x)')
ax[0,0].set_ylabel('sin(x)')
ax[0,1].plot(x,np.cos(x))
ax[1,0].plot(x,np.tan(x))
ax[1,1].plot(x,np.exp(x))
```
